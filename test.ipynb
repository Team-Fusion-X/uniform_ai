```python
import json
import pandas as pd

with open('json_data/2019_pass_or_fail.json', encoding='UTF8') as file:
    dataset = json.load(file)
```

```python
# DataFrame 생성
df = pd.DataFrame(dataset)

# 대학별 데이터 개수 확인
university_counts = df['대학명'].value_counts()

# 결과 출력
print(university_counts)
```

```python
# 상위 10개 데이터 출력
top_10 = university_counts.head(10)
print("상위 10개 데이터:")
print(top_10)

# 하위 10개 데이터 출력
bottom_10 = university_counts.tail(10)
print("\n하위 10개 데이터:")
print(bottom_10)
```
